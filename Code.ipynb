### data encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df.Grade = le.fit_transform(df.Grade)
df.ScholarType = le.fit_transform(df.ScholarType)
df.Direction = le.fit_transform(df.Direction)
df.Gender = le.fit_transform(df.Gender)
df.Medium = le.fit_transform(df.Medium)
df.CourseType = le.fit_transform(df.CourseType)
df.ProgramType = le.fit_transform(df.ProgramType)
```

```python
df = df.fillna(0.00)
```

```python
df
```

### data training and testing

```python
from sklearn.model_selection import train_test_split
x = df.iloc[:,3:8]
y = df.iloc[:,2]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

### feature scaling

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train_ = ss.fit_transform(x_train)
x_test_ = ss.transform(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
x_train_ = mms.fit_transform(x_train)
x_test_ = mms.transform(x_test)
```

```python
x_train_
```

```python
print(x_train_.shape)
print(x_train.shape)
```

```python
x_test_
```

```python
from sklearn.naive_bayes import GaussianNB
df = GaussianNB()
df.fit(x_train_,y_train)
y_test_ = df.predict(x_test_)
```

```python
from sklearn.metrics import accuracy_score,f1_score,r2_score,classification_report
print(accuracy_score(y_test,y_test_))
print(f1_score(y_test,y_test_,average = 'micro'))
print(r2_score(y_test,y_test_))
print(classification_report(y_test,y_test_))
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train_,y_train)
y_test_ = svc.predict(x_test_)
```

```python
print(accuracy_score(y_test,y_test_))
print(f1_score(y_test,y_test_,average = 'micro'))
print(r2_score(y_test,y_test_))
print(classification_report(y_test,y_test_))
```

```python
from sklearn.ensemble import BaggingClassifier
bg = BaggingClassifier()
bg.fit(x_train_,y_train)
y_test_ = bg.predict(x_test_)
```

```python
print(accuracy_score(y_test,y_test_))
print(f1_score(y_test,y_test_,average = 'micro'))
print(r2_score(y_test,y_test_))
print(classification_report(y_test,y_test_))
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier()
ada.fit(x_train_,y_train)
y_test_ = ada.predict(x_test_)
```

```python
print(accuracy_score(y_test,y_test_))
print(f1_score(y_test,y_test_,average = 'micro'))
print(r2_score(y_test,y_test_))
print(classification_report(y_test,y_test_))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train_,y_train)
y_test_ = dtc.predict(x_test_)
```

```python
print(accuracy_score(y_test,y_test_))
print(f1_score(y_test,y_test_,average = 'micro'))
print(r2_score(y_test,y_test_))
print(classification_report(y_test,y_test_))
```
